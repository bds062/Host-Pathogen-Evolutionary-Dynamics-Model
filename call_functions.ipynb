```python
from model_functions import Model, RasterModel, GraphModel
import numpy as np
import time
```

```python
param1_values = np.linspace(0, 365, 25)
param2_values = np.linspace(0, 365, 25)

model = Model(maturity=8/365)
raster_model = RasterModel(model, 'infected_offset', 'germination_offset', 
                           param1_values, param2_values)

print(model.maturity*365)

start_time = time.time()
raster_model.raster()  # Uses all 8 cores automatically
end_time = time.time()
print(f"\nTotal execution time: {end_time - start_time:.2f} seconds")

raster_model.plot_heatmap()
```
