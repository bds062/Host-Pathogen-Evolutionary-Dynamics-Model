```python
from model_functions import Model, RasterModel, GraphModel
import numpy as np
import time
```

```python
param1_values = np.linspace(300, 365, 8) #vertical axis
param2_values = np.linspace(250, 365, 10) #horizontal axis



model = Model(maturity=4/365, t = (0, 365 * 400), germination_time = 302.5, Bj = 0.2 / 365, infected_time=182.5, infected_time2=182.5)
# graph_model = GraphModel(model)
raster_model = RasterModel(model, 'infected_time', 'infected_time2', param1_values, param2_values)

print("Maturation rate:", model.maturity*365)
print("BetaJ: ", model.Bj*365)
print("BetaA: ", model.Bv*365)


start_time = time.time()
# graph_model.graph()
# print("Evaluation (genotype1/genotype1+2): ", model.evaluate(model.solution))
# raster_model.sweep_1d(param_num=1)
# raster_model.plot_1d()
raster_model.raster()  # Uses all 8 cores automatically
end_time = time.time()
print(f"\nTotal execution time: {end_time - start_time:.2f} seconds")

raster_model.plot_heatmap()
```
