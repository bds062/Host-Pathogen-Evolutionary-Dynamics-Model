```python
import sys
sys.path.append('../src')

from model import Model
from plot import RasterModel, GraphModel
import numpy as np
import time
```

```python
param1_values = np.linspace(0, 365, 15) #vertical axis
param2_values = np.linspace(0, 365, 15) #horizontal axis



model = Model(X_0 = (5, 0, 0, 5, 0, 0, 0, 0, 0, 0), t = (0, 365 * 10), germination_time = 212.5, Bj = 0.2 / 365, infected_time=182.5, infected_time2=182.5)
# graph_model = GraphModel(model)
raster_model = RasterModel(model, 'infected_time', 'infected_time2', param1_values, param2_values)

# print("Maturation rate:", model.maturity*365)
print("BetaJ: ", model.Bj*365)
print("BetaA: ", model.Bv*365)


start_time = time.time()
raster_model.plot_fig1(
    germination_values=np.linspace(1, 364, 15),
    maturity_values=[2/365, 8/365, 16/365],
    bj_values=[0.1/365, 0.2/365, 0.3/365],
    maturity_labels=["Slow (2 yr$^{-1}$)", "Default (8 yr$^{-1}$)", "Fast (16 yr$^{-1}$)"],
    bj_labels=[r"$\beta_J$ = 0.10 yr$^{-1}$", r"$\beta_J$ = 0.20 yr$^{-1}$", r"$\beta_J$ = 0.30 yr$^{-1}$"],
    save_path="../../PubFigs/fig1.pdf",
)
end_time = time.time()
print(f"\nTotal execution time: {end_time - start_time:.2f} seconds")
```
